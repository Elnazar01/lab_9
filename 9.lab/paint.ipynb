```python
import pygame
import random

# Инициализация Pygame
pygame.init()

# Создание холста
screen = pygame.display.set_mode((900, 700))
screen.fill((255, 255, 255))

# Установка заголовка окна
pygame.display.set_caption('GFG Paint')

draw_on = False
last_pos = (0, 0)

# Радиус кисти
radius = 5 

# Цвета
WHITE = (255, 255, 255)
RED = (255, 0, 0)
YELLOW = (255, 255, 0)
GREEN = (102, 204, 0)
BLUE = (51, 51, 255)
BLACK = (0, 0, 0)
PINK = (255, 0, 255)

# Цвет по умолчанию
color = BLACK  

# Отрисовка палитры цветов
pygame.draw.rect(screen, RED, (0, 50, 20, 20))
pygame.draw.rect(screen, YELLOW, (0, 70, 20, 20))
pygame.draw.rect(screen, GREEN, (20, 50, 20, 20))
pygame.draw.rect(screen, BLUE, (20, 70, 20, 20))
pygame.draw.rect(screen, BLACK, (0, 90, 20, 20))
pygame.draw.rect(screen, PINK, (20, 90, 20, 20))

# Попытка загрузить изображение ластика
try:
    erasor = pygame.transform.scale(pygame.image.load("racer_files01/erasor.jpg"), (40, 40))
    screen.blit(erasor, [0, 110])
except FileNotFoundError:
    print("Файл 'erasor.jpg' не найден. Убедитесь, что он находится в нужной папке.") 

# Цвет ластика (на белом фоне используем чёрный)
eraser_color = BLACK  

# Функция для плавной линии между двумя точками
def roundline(canvas, color, start, end, radius=1):
    Xaxis = end[0] - start[0]
    Yaxis = end[1] - start[1]
    dist = max(abs(Xaxis), abs(Yaxis))
    for i in range(dist):
        x = int(start[0] + float(i) / dist * Xaxis)
        y = int(start[1] + float(i) / dist * Yaxis)
        pygame.draw.circle(canvas, color, (x, y), radius)  

# Функция для стирания (рисуем цветом фона)
def erase(spot):
    pygame.draw.circle(screen, eraser_color, spot, radius)

try:
    while True:
        e = pygame.event.wait()

        if e.type == pygame.QUIT:
            raise StopIteration  

        if e.type == pygame.MOUSEBUTTONDOWN:
            spot = pygame.mouse.get_pos()
            # Выбор цвета по координатам мыши
            if spot[0] < 20 and 50 < spot[1] < 70:
                color = RED
            elif 20 < spot[0] < 40 and 50 < spot[1] < 70:
                color = GREEN
            elif spot[0] < 20 and 70 < spot[1] < 90:
                color = YELLOW
            elif 20 < spot[0] < 40 and 70 < spot[1] < 90:
                color = BLUE
            elif spot[0] < 20 and 90 < spot[1] < 110:
                color = BLACK
            elif 20 < spot[0] < 40 and 90 < spot[1] < 110:
                color = PINK
            elif spot[0] < 40 and 110 < spot[1] < 150:
                color = WHITE  # Ластик (белый цвет)

            # Начинаем рисовать, если клик за пределами меню
            if spot[0] > 60:
                pygame.draw.circle(screen, color, e.pos, radius)
            draw_on = True

        if e.type == pygame.MOUSEBUTTONUP:
            draw_on = False          

        if e.type == pygame.MOUSEMOTION:
            spot = pygame.mouse.get_pos()
            if draw_on and spot[0] > 60:
                pygame.draw.circle(screen, color, e.pos, radius)
                roundline(screen, color, e.pos, last_pos, radius)
            last_pos = e.pos

        if e.type == pygame.KEYDOWN:
            spot = pygame.mouse.get_pos()

            # Прямоугольник — клавиша R
            if e.key == pygame.K_r:
                pygame.draw.rect(screen, color, (spot[0], spot[1], 100, 150))

            # Круг — клавиша C
            elif e.key == pygame.K_c:
                pygame.draw.circle(screen, color, (spot[0], spot[1]), 50)

            # Стирание — клавиша E
            elif e.key == pygame.K_e:
                erase(spot)

            # Квадрат — клавиша S
            elif e.key == pygame.K_s:
                size = 100
                pygame.draw.rect(screen, color, (spot[0], spot[1], size, size))

            # Прямоугольный треугольник — клавиша T
            elif e.key == pygame.K_t:
                points = [
                    (spot[0], spot[1]),                # вершина
                    (spot[0] + 100, spot[1]),          # основание
                    (spot[0], spot[1] - 100)           # высота
                ]
                pygame.draw.polygon(screen, color, points)

            # Равносторонний треугольник — клавиша Q
            elif e.key == pygame.K_q:
                side = 100
                height = int((3 ** 0.5 / 2) * side)
                points = [
                    (spot[0], spot[1] - height // 2),          # верх
                    (spot[0] - side // 2, spot[1] + height // 2),  # низ слева
                    (spot[0] + side // 2, spot[1] + height // 2)   # низ справа
                ]
                pygame.draw.polygon(screen, color, points)

            # Ромб — клавиша H
            elif e.key == pygame.K_h:
                width = 80
                height = 100
                points = [
                    (spot[0], spot[1] - height // 2),   # верх
                    (spot[0] - width // 2, spot[1]),    # лево
                    (spot[0], spot[1] + height // 2),   # низ
                    (spot[0] + width // 2, spot[1])     # право
                ]
                pygame.draw.polygon(screen, color, points)

        # Обновление экрана
        pygame.display.flip()

except StopIteration:
    pass

# Завершение работы
pygame.quit()
```
